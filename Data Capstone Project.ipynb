```python
# how many country have republic in their full name
```

```python
count = 0
def counting_republic(long_name): 
    global count 
    if 'republic' in long_name.lower(): 
        count += 1
    return long_name

countries['country_long'] = countries['country_long'].apply(counting_republic)
print(count)
```

```python
africa_count = countries[countries['continent'] == 'Africa']
print(africa_count)
```
