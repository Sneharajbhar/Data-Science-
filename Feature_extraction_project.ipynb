```python
def extract_episode(title): 
    check = False
    ep_data = ""
    for i in title:
        if i == ')': 
            check = False
            return ep_data 
            break 
        if check == True:
            ep_data += i 
        if i == '(': 
            check = True
```

```python
df["No of episodes"] = df["Title"].apply(extract_episode)
```

```python
def extraction_time(title): 
    time_data = ""
    check = False 
    for i in range(len(title)):
        if title[i] == ')': 
            for j in range(i+1,i+20): 
                time_data += title[j]
                
    return time_data
```

```python
df["Timestamp"] = df["Title"].apply(extraction_time)
```

```python
from dateutil.relativedelta import relativedelta 
from datetime import datetime 

def calculate_total_months(peroid): 
    try: 
        start_str,end_str = peroid.split(' - ')
        start_date = datetime.strptime(start_str,'%b %Y')
        end_date = datetime.strptime(end_str,'%b %Y')
        r = relativedelta(end_date,start_date)
        return r.years * 12 +  r.months + 1 
    except: 
        return None 

df["Month"] = df["Timestamp"].apply(calculate_total_months)
```
