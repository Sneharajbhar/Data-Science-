# File Handling Simple project

```python
from pathlib import Path
import os

def show_file_names():
    path = Path('')
    items = list(path.rglob('*'))
    for i,items in enumerate(items):
        print(f"{i+1} : {items}")

def creating_a_file():
    try:
        show_file_names()
        file_name = input("Enter file name: ")
        
        new_path = Path(file_name)
        if not new_path.exists() and new_path.is_file():
            with open(new_path,"w") as file:
                data = input("Enter your data: ")
                file.write(data)
            print("File created successfully!")
        else:
            print("This file already exist!")
    except Exception as err:
        print(f"An error occured as {err}")

def reading_a_file():
    try:
         show_file_names()
         to_read_file = input("Enter name of a file that you want to read:")
         path = Path(to_read_file)
         if path.exists() and path.is_file():
            with open(to_read_file) as file:
                data = file.read()
                print(data)
            print("File readed successfully!")
         else:
             print("File doesn't exits!")
    except Exception as err:
        print(f"An error occured as {err}")
        

def updating_a_file():
    try:
        show_file_names()
        to_update_file = input("Enter name of a file that you want to update:")
        path = Path(to_update_file)
        if path.exists() and path.is_file():
             print("Press 1 for updating file name")
             print("Press 2 for overwriting file content")
             print("Press 3 for appending/add data in a file") 
             reponse = int(input("Enter your choice: "))
             
             if reponse == 1:
                 new_file_name = input("Enter file name to update:")
                 new_path_file = Path(new_file_name)
                 path.rename(new_path_file)
                 print("File name updated successfully!")
             elif reponse == 2: 
                 with open(to_update_file,"w") as file:
                     data = input("Enter your data: ")
                     file.write(data)
                 print("File updated successfully!")
             elif reponse == 3:
                 with open(to_update_file,"a") as file:
                     data = input("Enter your data: ")
                     file.write(" "+data)
                     print("Data append successfully!") 
        else:
            print("No such File exits!")
    except Exception as err:
        print(f"An error occured as {err}")

def deletion_of_file():
     try:
         show_file_names()
         to_delete_file = input("Enter name of a file that you want to delete:")
         path = Path(to_delete_file)
         if path.exists() and path.is_file():
            os.remove(path)
            print("File deleted successfully!")
         else:
             print("No such File exits!")
     except Exception as err:
        print(f"An error occured as {err}")


print("Press 1 for creating a file")
print("Press 2 for reading a file")
print("Press 3 for updating a file")
print("Press 4 for deletion a file")

reponse = int(input("Enter your choice: "))

if reponse == 1:
    creating_a_file()
elif reponse == 2:
    reading_a_file() 
elif reponse == 3:
    updating_a_file()
elif reponse == 4:
    deletion_of_file()
    
 
 
    
```
