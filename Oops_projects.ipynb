```python
# JSON : JavaScript Object Notation
# importing packages and  modules
import json 
import random
import string
from pathlib import Path



class Bank:
    database = 'data.json'
    data = []
    try:
        if Path(database).exists():
            with open(database) as file:
                data = json.loads(file.read())
        else:
            print("No such file exist!")
    except Exception as err:
        print(f"An exception occured as {err}")
        
# update
    @staticmethod
    def __updateAccount():
        with open(Bank.database,'w') as file:
            file.write(json.dumps(Bank.data,indent = 4))


# account generate 
   
    def accountGenerate(self):
        alpha = random.choices(string.ascii_letters,k=3)
        num = random.choices(string.digits,k=3)
        special = random.choices("!@#$%^&*",k=1)
        id = alpha + num + special
        random.shuffle(id)
        return "".join(id)
        
        
    
# create account            
    def createAccount(self):
        info = {
            "name": input("Enter your name: "),
            "age": int(input("Enter your age: ")),
            "email": input("Enter your email: "),
            "pin": int(input("Enter you PIN: ")),
            "accountNo": Bank.__accountGenerate(),
            "balance":0,
        }

        if info['age'] < 18 or len(str(info['pin'])) != 4:
            print("Sorry!,You can't create your account.")
        else:
            print("Account has been created successfully!")
            for i in info:
                print(f"{i} : {info[i]}")
            print("Please note down your account")
            Bank.data.append(info)
            Bank.__updateAccount()

# deposit money
    def depositMoney(self):
        account_No = input("Enter your account no:")
        pin_input = int(input("Enter your pin:"))
        user_data = [i for i in Bank.data if i['accountNo'] == account_No and i['pin'] == pin_input]

        if not user_data:
            print("Oops!, Sorry not data is found.")
        else:
            amount = int(input("Enter amount that you want to deposit: "))
            if amount > 10000:
                print("Sorry!, The amount is too much you can deposit money below 10000.")
            elif amount <= 0:
                print("Sorry invalid amount can't be added in balance.")
            else:
                
                user_data[0]['balance'] += amount

                Bank.__updateAccount()
                print("Money deposited successfully!")

# withdraw money
    def withdrawMoney(self):
        account_No = input("Enter your account no:")
        pin_input =int(input("Enter your pin:"))
        user_data = [i for i in Bank.data if i['accountNo'] == account_No and i['pin'] == pin_input]

        if not user_data:
            print("Oops!, Sorry no data is found.")
        else:
            amount = int(input("Enter amount that you want to withdraw: "))
            if user_data[0]['balance'] <  amount:
                print("Sorry!, Your not have sufficent balance.")
            else:
               
                user_data[0]['balance'] -= amount

                Bank.__updateAccount()
                print("Amount withdraw successfully!")

# show details

    def showDetails(self):
        account_No = input("Enter your account no:")
        pin_input = int(input("Enter your pin:"))

        user_data = [i for i in Bank.data if i['accountNo'] == account_No and i['pin'] == pin_input]
        if not user_data:
            print("Not! found any details.")
            
        else:
           print("Your Details:") 
           for i in user_data[0]: 
               print(f"{i} : {user_data[0][i]}")

# updating the details
    def updateDetails(self):
        account_No = input("Enter your account no:")
        pin_input = int(input("Enter your pin:"))
        user_data = [i for i in Bank.data if i['accountNo'] == account_No and i['pin'] == pin_input]

        if not user_data:
            print("No such user found!")
        else:
            print("You can't change the age, account no")

            print("Fill the details for change or leave it empty if no change")

            new_data = {
                "name": input("Enter new name/press enter for blank: "),
                "email": input("Enter new name/press enter for email: "),
                "pin": input("Enter new name/press enter for pin: "),
            }

            if new_data["name"] == "":
                new_data["name"] = user_data[0]["name"]
            if new_data["email"] == "":
                new_data["email"] = user_data[0]["email"]
            if new_data["pin"] == "":
                new_data["pin"] = user_data[0]["pin"]
            new_data['age'] = user_data[0]['age']
            new_data['accountNo'] = user_data[0]['accountNo']
            new_data['balance'] = user_data[0]['balance']

            if type(new_data['pin']) == str:
                new_data['pin'] = int(new_data['pin'])

            for i in new_data:
                if new_data[i] == user_data[0][i]:
                    continue 
                else:
                    user_data[0][i] = new_data[i]
            Bank.__updateAccount()
            print("Details updated successfully!")
            
# delete Account

    def deleteAccount(self):
        account_No = input("Enter your account no:")
        pin_input = int(input("Enter your pin:"))

        user_data = [i for i in Bank.data if i['accountNo'] == account_No and i['pin'] == pin_input]

        if not user_data:
            print("No such data exist!")
        else:
            check = input("Press y if you actually want to delete your account otherwise n: ")
            if check in ['n','N']:
                pass
            elif check in ['y','Y']:
                index = Bank.data.index(user_data[0])
                Bank.data.pop(index)
                print("Account deleted successfully!")
                Bank.__updateAccount()
            else:
                print("You enter wrong option!, please try again.")
            
            
# obj of a class
user = Bank()   

# options
print("Press 1 for creating an account")
print("Press 2 for depositing money in the bank")
print("Press 3 for withdrawing the money")
print("Press 4 for details")
print("Press 5 for updating the details")
print("Press 6 for deleting your account")

response = int(input("Enter your choice: "))

# condition match
if response == 1:
    user.createAccount()
elif response == 2:
    user.depositMoney()
elif response == 3:
    user.withdrawMoney()
elif response == 4:
    user.showDetails()
elif response == 5:
    user.updateDetails()
elif response == 6:
    user.deleteAccount()
else:
    print("Invaild choice!")
```
