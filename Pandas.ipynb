# Pandas Series

A Series is one-demensional labeled array capable of holding any data type.The axis labels are collectively called the index.

## Creating Series

There are multiple ways to create a pandas Series:


```python
import numpy as np
import pandas as pd 
```

```python
lables = ['a','b','c']
my_list = [10,20,30]

arr = np.array([10,20,30])

dict1 = {'a':10,'b':20,'c':30}

pd.Series(my_list)
```

```python
pd.Series(my_list,index=lables)
```

```python
pd.Series(arr)
```

```python
pd.Series(dict1)
```

### Merging 2 DataFrame

```python
employees = pd.DataFrame(
    {
        'emp_id' : [1,2,3,4,5], 
        'name' : ['Sneha','Manvi','Shipra','Aditi','khushi'], 
        'department' : ['HR','IT','Finance','IT','HR'], 
    }
)

salaries = pd.DataFrame(
    {
        'emp_id' : [1,2,3,6,7], 
        'salary' : [600000,800000,650000,700000,900000], 
        'bonus' : [5000,10000,70000,80000,12000]
    }
)

```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on="emp_id",how = "inner")
```

```python
pd.merge(employees,salaries,on="emp_id",how = "outer")
```

```python
pd.merge(employees,salaries,on="emp_id",how = "left")
```

```python
pd.merge(employees,salaries,on="emp_id",how = "right")
```

**GroupBY**

```python
data = {
    'Category' : ['A','B','A','B','A','B','A','B'], 
    'Store' : ['S1','S1','S2','S2','S1','S2','S2','S1'], 
    'Sales' : [100,200,150,250,120,180,200,300], 
    'Quantity' : [10,15,12,18,8,20,15,25], 
    'Date' : pd.date_range('2023-01-01',periods=8)
} 

df = pd.DataFrame(data)
```

```python
df
```

```python
# group by category and calculate the sum of sales

cat = df.groupby('Category')
for i,v in cat:
    print(i) 
    print(v)
```

```python
cat = df.groupby('Category')['Sales'].sum()
print(cat)
```

```python
store_sum = df.groupby('Store')['Sales'].sum()
print(store_sum)
```

```python
# Group by multiple columns
cat = df.groupby(['Category','Store'])['Sales'].sum()
print(cat)
```

**Pivot Tables**

```python

data = {
    'Date' : pd.date_range('2023-01-01',periods=20), 
    'Product' : ['A','B','C','D','E']*4, 
    'Region' : ['East','West','North','South']*5,
    'Units' : np.random.randint(10,100,20), 
    'Sales' : np.random.randint(100,1000,20), 
    'Rep' : ['Sneha','Manvi','Riya','Shipra']*5
} 
df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
```

```python
df
```

```python
pd.pivot_table(df,values = 'Sales',index = 'Region',columns='Product',aggfunc='median')
```

```python
pd.pivot_table(df,values = ['Sales','Units'],index = 'Region',columns='Product',aggfunc='median')
```

**Cross Tabs** 

```python
pd.crosstab(df['Region'],df['Product'])
```
